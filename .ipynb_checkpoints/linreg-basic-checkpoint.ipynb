## Linear regression

In this note we will describe the linear regression algorithm. There are several ways to do this, so lets first set up the problem. Suppose we are given a dataset $\{x_i, y_i\}_{i\in I}$ and want to fit this via a linear relationship. In other words, we are supposing that for some parameters $\theta_0, \theta_1$, the data set is sampled from the distribution
        $$ y_i \sim\mathcal{N}(\theta_1 x_i + \theta_0, \sigma^2) $$
where $\sigma^2$ is a fixed parameter.

Our goal is to determine the parameters $\theta_i$ for this data, or at least, determine the posterior distribution for the parameters, $p(\theta|y,x)$.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

The first choice for the job is a maximal likelihood estimate (MLE) approach. The likelihood is given by $p(y|\theta, x)$, i.e. the probability of seeing the data given the current parameters. MLE seeks to find the parameters $\theta_i$ that maximizes this distribution,

$$ \theta_\text{MLE} = \underset{\theta}{\operatorname{argmax}} p(y|\theta, x) $$

As we are looking at a Gaussian distribution $y_i\sim\mathcal{N}(\theta_1 x+\theta_0, \sigma^2)$, we have

$$ p(y|\theta,x)=\prod_{i=1}^n p(y_i|\theta, x_i) = \prod_{i=1}^n \frac{1}{\sqrt{2\pi\sigma^2}}
\exp\left(-\frac{(y_i-(\theta_1 x_i+\theta_0))^2}{2\sigma^2}\right) $$

and so taking log-likelihoods, we want to maximize

$$ \mathcal{L}=\log p(y|\theta,x) = \text{const} -\frac{1}{2\sigma^2}\sum^n_{i=1}(y_i-(\theta_1 x_i+\theta_0))^2 $$

```python
# Generate sample data
n = 70
slope = 2
bias = 1
noise = 0.7
errors = noise * np.random.randn(n)
# x, y pairs
x = np.random.uniform(0, 3, n)
y = slope * x + bias + errors

# Plot sample data
plt.figure(figsize=(10,4))
plt.scatter(x, y)
x_line = np.linspace(0, 3)
plt.plot(x_line, x_line * slope + bias)
plt.show()
```

Computing the log-gradients,
$$ \frac{\partial\mathcal{L}}{\partial\theta_0} = \frac{1}{\sigma^2}\sum_{i=1}^n(y_i-(\theta_1 x_i+\theta_0)) $$
and
$$ \frac{\partial\mathcal{L}}{\partial\theta_1} = \frac{1}{\sigma^2}\sum_{i=1}^n(y_i-(\theta_1 x_i+\theta_0))x_i $$
setting to zero and solving, we get
$$ \theta_{1,\text{MLE}}=\frac{\sum_{i=1}^n (x_i-\bar{x})(y_i-\bar{y})}{\sum_{i=1}^n (x_i-\bar{x})^2} \text{ and }
\theta_{0,\text{MLE}}=\bar{y}-\theta_{1,\text{MLE}}\bar{x}$$
where $\bar{x},\bar{y}$ are the respective means.

This can be directly used to find **least squares best fit line**.

```python
# MLE regression
x_mean = np.mean(x)
y_mean = np.mean(y)

# MLE parameters
theta_1_mle = np.sum((x-x_mean)*(y-y_mean)) / np.sum((x-x_mean)**2)
theta_0_mle = y_mean - theta_1_mle * x_mean

# Plot guessed line
plt.figure(figsize=(10,4))
plt.scatter(x,y)
x_line = np.linspace(0, 3)
plt.plot(x_line, theta_1_mle * x_line + theta_0_mle, '-r', label="MLE line")
plt.plot(x_line, slope * x_line + bias,label="True line")
plt.legend()
plt.show()
print("Guessed theta_0: ", theta_0_mle)
print("Guessed theta_1: ", theta_1_mle)
```

Pretty good!
