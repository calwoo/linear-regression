## Linear regression - descent

Sometimes an exact formula for the parameters doesn't cut it (or is impossible to attain analytically), so we resort to approximation methods. The most popular of such is hill-climbing, or **gradient descent**. To do this, we convert the previous note to the language of matrices.

As before, we are computing the log-likelihood
$$ \mathcal{L}=\log p(y|\theta,x) = \text{const} -\frac{1}{2\sigma^2}\sum^n_{i=1}(y_i-(\theta_1 x_i+\theta_0))^2 $$
We can let

* $\Theta=[\theta_0, \theta_1] $
* $X = [\{(1,x_i)\}_{i\in I}]$
* $Y = [\{y_i\}_{i\in I}]$

Then with this, we are thus trying to find the MLE parameters

$$ \Theta_\text{MLE} = \underset{\theta}{\operatorname{argmin}} \left\Vert Y-X\Theta\right\Vert^2 $$
The vectorized matrix derivative is
$$ \frac{\partial\mathcal{L}}{\partial\Theta} = -2X^T(Y-X\Theta) $$

```python
# Imports and data
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Generate sample data
n = 40
slope = 3
bias = 2
noise = 0.7
errors = noise * np.random.randn(n)
# x, y pairs
x = np.random.uniform(0, 3, n)
y = slope * x + bias + errors

# Plot sample data
plt.figure(figsize=(10,4))
plt.scatter(x, y)
x_line = np.linspace(0, 3)
plt.plot(x_line, x_line * slope + bias)
plt.show()
```

Given a learning rate $\eta$, we can step towards an optimal set of parameters by continually computing the model predictions, computing the loss, computing the gradients, updating the parameters by the gradient descent step, and then repeating. The **gradient descent step** in this case is the update rule

$$ \Theta^\text{new} \leftarrow \Theta - \eta\frac{\partial\mathcal{L}}{\partial\Theta} $$

This learning rate is a **hyperparameter** and needs to well-tuned for the gradient descent to work well. This is a simple model, but more advanced models with complicated loss functions might require tuning at each step for good convergence behaviors.

```python
# Gradient descent for linear regression
X = np.c_[np.ones(n), x]
Y = y
Theta = np.random.randn(2)
learning_rate = 0.001

# NOTE: The @ symbol is matrix multiplication. Great!
```

```python
# Gradient descent update
def update(X, Y, Theta, learning_rate):
    grad = -2 * X.T @ (Y - (X @ Theta))
    Theta -= learning_rate * grad
    return Theta
```

Now to run the gradient descent algorithm to get a good set of parameters.

```python
losses = [] # Container for loss
params = [] # Container for parameters

for _ in range(50):
    Theta = update(X, Y, Theta, learning_rate)
    losses.append(np.sum((Y-X@Theta)**2))
    params.append(Theta)
    
print("Guessed theta_0: ", Theta[0])
print("Guessed theta_1: ", Theta[1])
```

How did we do?

```python
# Plot guessed line
plt.figure(figsize=(10,4))
plt.scatter(x,y)
x_line = np.linspace(0, 3)
plt.plot(x_line, Theta[1] * x_line + Theta[0], '-r', label="GD line")
plt.plot(x_line, slope * x_line + bias,label="True line")
plt.legend()
plt.show()
```

Pretty well! Here's how the gradient changed on the loss surface over time.
